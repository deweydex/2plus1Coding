```
import random

def choose_word(category="general"):
    words = {"general": ["PYTHON", "CODING", "FUNCTION"],
             "animals": ["ELEPHANT", "GIRAFFE", "PENGUIN"]}
    return random.choice(words.get(category, words["general"]))

def display_word(word, guessed):
    return " ".join(letter if letter in guessed else "_" for letter in word)

def get_letter_guess(already_guessed):
    while True:
        guess = input("\nGuess a letter: ").upper().strip()
        if len(guess) != 1 or not guess.isalpha():
            print("Enter one letter.")
        elif guess in already_guessed:
            print(f"Already guessed {guess}.")
        else:
            return guess

# Demo
word = choose_word()
guessed = set()
wrong = 0

print("=" * 50)
print("  WORD GUESSING - Advanced Starter")
print("=" * 50)
print(f"\nGuess the {len(word)}-letter word! (Demo: 3 guesses)\n")

for _ in range(3):
    print(f"Word: {display_word(word, guessed)}")
    print(f"Wrong: {wrong}")
    
    guess = get_letter_guess(guessed)
    guessed.add(guess)
    
    if guess in word:
        print(f"✓ {guess} is in the word!")
    else:
        print(f"✗ {guess} is not in the word.")
        wrong += 1

print(f"\n{display_word(word, guessed)}")
print(f"Secret word: {word}")
print("\nNow add win/lose detection and full game loop!")
```
