```
def create_board():
    return ['1', '2', '3', '4', '5', '6', '7', '8', '9']

def display_board(board):
    print(f"\n {board[0]} | {board[1]} | {board[2]}")
    print("---|---|---")
    print(f" {board[3]} | {board[4]} | {board[5]}")
    print("---|---|---")
    print(f" {board[6]} | {board[7]} | {board[8]}\n")

def make_move(board, position, player):
    board[position - 1] = player

def get_player_move(player):
    while True:
        try:
            position = int(input(f"Player {player}, enter position (1-9): "))
            if 1 <= position <= 9:
                return position
            print("Enter 1-9.")
        except ValueError:
            print("Enter a number.")

# Demo
board = create_board()
current_player = 'X'
print("=" * 40)
print("  TIC-TAC-TOE - Advanced Starter")
print("=" * 40)
print("\nTry making 4 moves:\n")

for _ in range(4):
    display_board(board)
    pos = get_player_move(current_player)
    make_move(board, pos, current_player)
    current_player = 'O' if current_player == 'X' else 'X'

display_board(board)
print("\nNow add win detection and full game logic!")
```
